```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:,[3,4]].values
```

```python
import scipy.cluster.hierarchy as sch

dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters = 5, affinity='euclidean',linkage = 'ward')
test = hc.fit_predict(X)
```

```python
plt.scatter(X[test==0,0],X[test==0,1], s=100,c='red',label='Cluster 1')
plt.scatter(X[test==1,0],X[test==1,1],s=100,c='blue',label='Cluster 2')
plt.scatter(X[test==2,0],X[test==2,1],s=100,c='green',label='Cluster 3')
plt.scatter(X[test==3,0],X[test==3,1],s=100,c='yellow',label='Cluster 4')
plt.scatter(X[test==4,0],X[test==4,1],s=100,c='black',label='Cluster 5')
plt.legend()
```
